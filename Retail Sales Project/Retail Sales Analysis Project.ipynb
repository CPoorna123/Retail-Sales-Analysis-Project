# Retail Sales Analysis

```python
# Load Dataset
import pandas as pd
df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
df
```

```python
# Basic Info
print("First 5 rows:")
print(df.head())
```

```python
print("\n Shape of dataset:", df.shape)
```

```python
#  Convert Order Date & Ship Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df
```

```python
#  Create new columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['MonthName'] = df['Order Date'].dt.strftime('%b')
```

```python
df
```

```python
# Total Sales
df['TotalSales'] = df['Sales']
df
```

```python
# Total Profit
df['TotalProfit'] = df['Profit']
df
```

```python
# Clean missing values
df = df.dropna()
df
```

```python
# Monthly Sales Summary
monthly_sales = df.groupby(['Year', 'Month'])['TotalSales'].sum().reset_index()
monthly_sales
```

```python
#  Create Year-Month column for plotting & Power BI
monthly_sales['YearMonth'] = monthly_sales['Year'].astype(str) + "-" + monthly_sales['Month'].astype(str).str.zfill(2)
```

```python
# Save cleaned data
df.to_csv("superstore_clean.csv", index=False)
monthly_sales.to_csv("monthly_sales.csv", index=False)
```
